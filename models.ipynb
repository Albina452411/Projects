```python
from scipy.fft import fft
def to_fury(data):
  for i in range(len(data)):
    row = data.iloc[i].values
    fury_row  = fft(row)
    data.replace(row, fury_row)
  return data
train = pd.read_csv(io.BytesIO(uploaded_train['train.csv']))
train
```
