```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import numpy as np
```

```python
a=[]
with open("i-1572__i3d.txt") as file:
    for item in file:
            a.append(item.split(','))
a
```

```python
a = pd.DataFrame(a)
```

```python
b = a.iloc[3:,1:]
b
```

```python
a = b.iloc[3:,:-1]
```

```python
columns = b.iloc[0]
columns
```

```python
columns = columns.iloc[:-1]
```

```python
columns
```

```python
a.columns = columns
```

```python
a
```

```python
a=a.replace(to_replace='n/a', value='NaN', inplace=False, limit=None, regex=False)
a
```

```python
a_clean=a.dropna(axis=1, how='all')
a_clean
```

```python
a_cl=a_clean[['Object volume', 'Volume-equivalent sphere diameter']]
a_cl
```

```python
a_cl=a_cl.astype(float)
```

```python
bin_interval=[0,1.05,75]
bin_step=[0.05,1]
```

```python
l = list()
for i in range(len(bin_interval)-1):
    l.append(
        np.arange(
            bin_interval[i],
            bin_interval[i+1], 
            bin_step[i]
        )
    )
```

```python
l = np.concatenate(l)
l
```

```python
a_cl['Bins Name'] = pd.cut(a_cl['Volume-equivalent sphere diameter'], l, labels=l[1:]).astype(float)
```

```python
a_cl
```

```python
A=list(a_cl['Bins Name'])
```

```python
for i in range(len(A)):
    if A[i]>1:
        A[i]=1
A
```

```python
a_cl['Bins Name']=A
```

```python
a_cl['Bins Name']
```

```python
a_group =a_cl.groupby('Bins Name')
num = a_group['Bins Name'].unique()
```

```python
bbins = []
for i in range(len(bin_interval)-1):
    bbins.append(
        num[(num >= bin_interval[i]) & (num < bin_interval[i+1])].count()
    )

bbins
```

```python
widths = np.array([20,20,20])
step = 40
placement_steps = widths + step
```

```python
bins_calc_bounds = [1]
bins_placements_list = []
```

```python
for i in range(len(bbins)):
    
    bins_calc_bounds.append(
        bins_calc_bounds[-1] + placement_steps[i]*bbins[i]
    )

    bins_placements_list.append(
        np.arange(
            bins_calc_bounds[i],
            bins_calc_bounds[i+1], 
            placement_steps[i]
        )
    )
```

```python
bins_placements = np.concatenate(bins_placements_list)
```

```python
bins_placements
```

```python
bins_width = -np.concatenate(
    [
        np.full(bbins[0], widths[0]), 
        np.full(bbins[1], widths[1])
    ]
)
```

```python
aa=list(a_group['Object volume'])
aa
```

```python
fig, ax = plt.subplots(figsize=(35, 20), dpi=300)
plt.suptitle("Образец № I-1572 КпСТ=1.75% КпСТcav=0.80%",fontsize=40)

porosity_percentage = ((a_group['Object volume'].sum()/a_cl['Object volume'].sum())*100).to_numpy()

bar_numbers = ax.bar(
    bins_placements, 
    porosity_percentage, 
    width=bins_width,
    color='red',
    align='edge'
)

ax.bar_label(
    bar_numbers, 
    labels=a_group['Volume-equivalent sphere diameter'].count(), 
    padding=5, 
    rotation='vertical',
    fontsize=20
)

diameter = np.round(list(a_group.groups.keys()), 2)
ax.set_xticks(bins_placements, diameter, rotation=90)
ax.set_xlabel('Эффективный диаметр, мм', fontsize=30)
ax.set_xlim((0-step, np.ceil(bins_placements.max()+step)))
ax.set_yscale('log')
ax.tick_params(labelsize=25)
ax.set_ylabel('Доля расчетной пористости, %', fontsize=30);
plt.text(0.6,1, 'Количество объектов', fontsize = 50)
plt.arrow(60,0.1,40,0.8)
plt.text(300,10,'Вклад в общий объем пор', fontsize = 50)
plt.arrow(400,9,310,-2)
```
